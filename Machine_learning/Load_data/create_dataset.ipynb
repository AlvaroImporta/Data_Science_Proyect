```python
import requests
import pandas as pd

data_urls = [
  'https://raw.githubusercontent.com/alexdriedger/SlayTheSpireFightPredictor/master/out/1593042045/data_1593042198.json',
  'https://raw.githubusercontent.com/alexdriedger/SlayTheSpireFightPredictor/master/out/1593042045/data_1593042341.json',
  'https://raw.githubusercontent.com/alexdriedger/SlayTheSpireFightPredictor/master/out/1593042045/data_1593042469.json',
  'https://raw.githubusercontent.com/alexdriedger/SlayTheSpireFightPredictor/master/out/1593042045/data_1593042598.json',
  'https://raw.githubusercontent.com/alexdriedger/SlayTheSpireFightPredictor/master/out/1593042045/data_1593042728.json',
  'https://raw.githubusercontent.com/alexdriedger/SlayTheSpireFightPredictor/master/out/1593042045/data_1593042853.json',
  'https://raw.githubusercontent.com/alexdriedger/SlayTheSpireFightPredictor/master/out/1593042045/data_1593042968.json',
  'https://raw.githubusercontent.com/alexdriedger/SlayTheSpireFightPredictor/master/out/1593042045/data_1593043097.json',
  'https://raw.githubusercontent.com/alexdriedger/SlayTheSpireFightPredictor/master/out/1593042045/data_1593043113.json'
]

json_data = list()
data = []

for api_url in data_urls:
    print(f'Loading {api_url}')
    response = requests.get(api_url)
    print(len(response.json()))
    data+=response.json()
```

```python
character = []
cards = []
relics = []
ascension = []
floor = []
max_hp = []
entering_hp = []
damage_taken = []
potion_used = []
enemies = []


for i in range(len(data)):
    character.append(data[i]['character'])
    cards.append(data[i]['cards'])
    relics.append(data[i]['relics'])
    ascension.append(data[i]['ascension'])
    floor.append(data[i]['floor'])
    max_hp.append(data[i]['max_hp'])
    entering_hp.append(data[i]['entering_hp'])
    damage_taken.append(data[i]['damage_taken'])
    potion_used.append(data[i]['potion_used'])
    enemies.append(data[i]['enemies'])

dictionario = {

    'character': character,
    'cards': cards,
    'relics': relics,
    'ascension': ascension,
    'floor': floor,
    'max_hp': max_hp,
    'entering_hp': entering_hp,
    'damage_taken': damage_taken,
    'potion_used': potion_used,
    'enemies': enemies
}

df = pd.DataFrame(dictionario)
```

```python
df_ironclad = df[df['character'] == 'IRONCLAD']
df_ironclad.reset_index(drop=True, inplace=True)
df_silent = df[df['character'] == 'THE_SILENT']
df_silent.reset_index(drop=True, inplace=True)
df_defect = df[df['character'] == 'DEFECT']
df_defect.reset_index(drop=True, inplace=True)
df_watcher = df[df['character'] == 'WATCHER']
df_watcher.reset_index(drop=True, inplace=True)
```

```python
import ast
from collections import Counter

def CreateCardsColumns(dataframe):
    cards_lists = []
    for cards_list in dataframe.cards:
        cards_lists += cards_list

    total_cards = set(cards_lists)
    total_cards_ordered = list(sorted(total_cards))

    for card in total_cards_ordered:
        dataframe[card] = 0

def FillCardsColumns(dataframe):
    for i in range(len(dataframe)):
        cards_list = dataframe.cards[i]
        dict_row_cards  = Counter(cards_list)
        for card in dict_row_cards:
            dataframe.at[i, card] = dict_row_cards[card]
    dataframe.drop(columns=['cards'], inplace=True)

def CreateRelicsColumns(dataframe):
    relics_lists = []
    for relics_list in dataframe.relics:
        relics_lists += relics_list

    total_relics = set(relics_lists)
    total_relics_ordered = list(sorted(total_relics))

    for relic in total_relics_ordered:
        dataframe[relic] = 0

def FillRelicsColumns(dataframe):
    for i in range(len(dataframe)):
        relics_list = dataframe.relics[i]
        dict_row_relics  = Counter(relics_list)
        for relic in dict_row_relics:
            dataframe.at[i, relic] = dict_row_relics[relic]
    dataframe.drop(columns=['relics'], inplace=True)

def TransformEnemies(dataframe):
    enemies_list = []
    for enemy in dataframe.enemies:
        enemies_list.append(enemy)
    total_enemies= set(enemies_list)

    dict_enemies = {}
    for id,enemy in enumerate(total_enemies):
        dict_enemies[enemy] = id

    dataframe['enemies'] = dataframe['enemies'].map(dict_enemies)

def TransformAscensionFloor(dataframe):
    if dataframe.floor.max() != 100:
        dataframe["floor"] = dataframe["floor"] * 100 // 56

def DeleteColumns(dataframe):
    dataframe["level_asc_floor"] = dataframe.ascension * 100 + dataframe.floor
    dataframe.drop(columns=['ascension', 'floor', 'potion_used'], inplace=True)

def TransformCardsAndRelics(dataframe):
    CreateCardsColumns(dataframe)
    FillCardsColumns(dataframe)
    CreateRelicsColumns(dataframe)
    FillRelicsColumns(dataframe)
    TransformEnemies(dataframe)
    TransformAscensionFloor(dataframe)
    DeleteColumns(dataframe)
```
